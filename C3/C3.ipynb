# Q1

```python
def func(x,y):
    return y**2+1

def E_Method(xi,xf,yi,steps=100, xprnt = False):
    '''
    xi = initial x, xf = final x, yi = initial y, steps = number of steps to go for loop (default = 100), xprnt is a flag if one wants to
    print out the step size, h (default = False).
    '''
    x_vals = [xi]
    y_vals = [yi]

    h = (xf-xi)/steps #always want to reach the end of integral range not go past it and approx
    print(f"The step size is {h}.")
    
    for i in range(steps):
        y_new = yi + h * func(x=xi,y=yi)
        x_new = xi + h

        if xprnt == True:
            print(x_new)
            
        yi = y_new
        xi = x_new
        

        x_vals.append(xi)
        y_vals.append(yi)

    return np.array(x_vals), np.array(y_vals)

def RK4(xi, xf, yi, steps=100, xprnt = False):
    '''
    xi = initial x, xf = final x, yi = initial y, steps = number of steps to go for loop (default = 100), xprnt is a flag if one wants to
    print out the step size, h (default = False).    '''
    
    x_vals = [xi]
    y_vals = [yi]

    h = (xf-xi)/steps #always want to reach the end of integral range not go past it and approx
    print(f"The step size is {h}.")
    
    for i in range(steps):
        k1 = func(xi,yi)
        k2 = func(xi+(h/2), yi+k1*(h/2))
        k3 = func(xi+(h/2),yi+k2*(h/2))
        k4 = func(xi+h,yi+k3*h)
        
        y_new = yi + (h/6)*(k1+2*k2+2*k3+k4)
        x_new = xi + h

        if xprnt == True:
            print(x_new)
            
        yi = y_new
        xi = x_new
        

        x_vals.append(xi)
        y_vals.append(yi)

    return np.array(x_vals), np.array(y_vals)
```

## Q1a

```python
x_vals, y_vals = E_Method(xi=0,xf=1.5,yi=0)
plt.plot(x_vals,y_vals, "go",markersize=4,label="Euler approx")
plt.plot(x_vals,np.tan(x_vals),label = "tan(x)")
plt.legend()
plt.savefig("E_basic",dpi=300)
```

```python
x_vals, y_vals = RK4(xi=0,xf=1.5,yi=0)
plt.plot(x_vals,y_vals, "ko",markersize=4,label="RK4 approx")
plt.plot(x_vals,np.tan(x_vals),label = "tan(x)")
plt.legend()
plt.savefig("RK4_basic",dpi=300)
```

### Testing Breaks with previous function

```python
def E_Method_break(xi,yi,h=0.01,steps=150, xprnt = False):
    '''
    h = step size (default = 0.01), steps = number of steps to go for loop (default = 150). 
    '''
    x_vals = [xi]
    y_vals = [yi]
    
    for i in range(steps):
        y_new = yi + h * func(x=xi,y=yi)
        x_new = xi + h

        if xprnt == True:
            print(x_new)
            
        yi = y_new
        xi = x_new
        

        x_vals.append(xi)
        y_vals.append(yi)

    return np.array(x_vals), np.array(y_vals)
```

```python
x_vals, y_vals = E_Method_break(xi=0,yi=0,steps=170,xprnt=True)
plt.plot(x_vals,y_vals, label="Euler approx")
plt.plot(x_vals,np.tan(x_vals),label = "tan(x)")
plt.legend()
```

```python
x_vals, y_vals = E_Method_break(xi=0,yi=0,steps=170)
plt.plot(x_vals,y_vals, label="Euler approx")
plt.plot(x_vals,np.tan(x_vals),label = "tan(x)")
plt.legend()
#plt.xlim(1.64,1.70)
plt.ylim(-100,1000)
#actual function still retains its shape, but the integrator shoots straight up bc the slope goes to infinity 
#slope goes to infinity at pi/2, integrator will try and go past it, but it will screw up in the end. 
```
