# Q1

```python
def func(x,y):
    return y**2+1

def E_Method(xi,xf,yi,steps=100, xprnt = False):
    '''
    xi = initial x, xf = final x, yi = initial y, steps = number of steps to go for loop (default = 100), xprnt is a flag if one wants to
    print out the step size, h (default = False).
    '''
    x_vals = [xi]
    y_vals = [yi]

    h = (xf-xi)/steps #always want to reach the end of integral range not go past it and approx
    print(f"The step size is {h}.")
    
    for i in range(steps):
        y_new = yi + h * func(x=xi,y=yi)
        x_new = xi + h

        if xprnt == True:
            print(x_new)
            
        yi = y_new
        xi = x_new
        

        x_vals.append(xi)
        y_vals.append(yi)

    return np.array(x_vals), np.array(y_vals)

def RK4(xi, xf, yi, steps=100, xprnt = False):
    '''
    xi = initial x, xf = final x, yi = initial y, steps = number of steps to go for loop (default = 100), xprnt is a flag if one wants to
    print out the step size, h (default = False).    '''
    
    x_vals = [xi]
    y_vals = [yi]

    h = (xf-xi)/steps #always want to reach the end of integral range not go past it and approx
    print(f"The step size is {h}.")
    
    for i in range(steps):
        k1 = func(xi,yi)
        k2 = func(xi+(h/2), yi+k1*(h/2))
        k3 = func(xi+(h/2),yi+k2*(h/2))
        k4 = func(xi+h,yi+k3*h)
        
        y_new = yi + (h/6)*(k1+2*k2+2*k3+k4)
        x_new = xi + h

        if xprnt == True:
            print(x_new)
            
        yi = y_new
        xi = x_new
        

        x_vals.append(xi)
        y_vals.append(yi)

    return np.array(x_vals), np.array(y_vals)
```

## Q1d
$e \propto \Delta x^{P}$ where P is the "order" of the method (will be the slope in the plot)

```python
#relative error for comparison (Euler only)
ran_con = [15,150,1500,15000,150000]

#highest-res step sizes - array of xs and ys
xE_ref, yE_ref = E_Method(0, 1.5, 0, ran_con[-1])

#now to compute the fractional difference:
frac_E = []
xE_vals = []

#need to loop over the smaller-res step sizes and then compute the fractional difference
for i in ran_con[:-1]:
    #print(i)
    x_vals, y_vals = E_Method(xi=0,xf=1.5,yi=0,steps=i)
    xE_vals.append(x_vals)
    
    #fractional difference from highest-res reference (last point)
    frac_E.append(abs(y_vals[-1] - yE_ref[-1]) / abs(yE_ref[-1]))
    

#highest-res step sizes - array of xs and ys
xR_ref, yR_ref = RK4(0, 1.5, 0, ran_con[-1])

frac_R = []
xR_vals = []
#need to loop over the smaller-res step sizes and then compute the fractional difference
for i in ran_con[:-1]:
    #print(i)
    x_vals, y_vals = RK4(xi=0,xf=1.5,yi=0,steps=i)
    xR_vals.append(x_vals)
    
    #fractional difference from highest-res reference (last point)
    frac_R.append(abs(y_vals[-1] - yR_ref[-1]) / abs(yR_ref[-1]))

plt.loglog(ran_con[:-1], frac_E, 'o-', label='Euler') #want to plot against delta x
plt.loglog(ran_con[:-1], frac_R, 'o-', label='RK4') #want to plot against delta x

plt.xlabel("Number of steps")
plt.ylabel("Fractional difference")
plt.legend()
plt.savefig("Convergence.png",dpi=300)
plt.show()
```

```python
def vel_dist(v,dv):
    '''
    For Hydrogen atoms. v is the range of velocities, dv = step size
    '''
    k = 1.380649e-23 #m2 kg s-2 K-1
    m = 1.6735575e-27 #kg
    T = 10000 #K

    return ((m/(2*np.pi*k*T))**(3/2)*4*np.pi*(v**2)*np.exp(-(m*v**2)/(2*k*T))) * dv

def v_p():
    k = 1.380649e-23 #m2 kg s-2 K-1
    m = 1.6735575e-27 #kg
    T = 10000 #K
    return np.sqrt((2*k*T)/m)

def v_avg():
    return (2/np.sqrt(np.pi))*v_p()

def v_RMS():
    return np.sqrt(3/2)*v_p()
```
