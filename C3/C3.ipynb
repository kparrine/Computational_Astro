```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x,y):
    return y**2+1

def E_Method(xi, yi, h=0.01, steps=100):
    '''
    h = step size (default = 0.01), steps = number of steps to go for loop (default = 100). 
    '''
    x_vals = [xi]
    y_vals = [yi]

    for i in range(steps):
        y_new = yi + h * func(x=0,y=yi)
        x_new = xi + h

        yi = y_new
        xi = x_new

        x_vals.append(xi)
        y_vals.append(yi)

    return np.array(x_vals), np.array(y_vals)
```

```python
x_vals, y_vals = E_Method(xi=0,yi=0)
plt.plot(x_vals,y_vals, label="Euler approx")
plt.plot(x_vals,np.tan(x_vals),label = "tan(x)")
plt.legend()
```

```python
def RK4(xi, yi, h=0.01, steps=100):
    '''
    h = step size (default = 0.01), steps = number of steps to go for loop (default = 100). 
    '''
    
    x_vals = [xi]
    y_vals = [yi]

    for i in range(steps):
        k1 = func(xi,yi)
        k2 = func(xi+(h/2), yi+k1*(h/2))
        k3 = func(xi+(h/2),yi+k2*(h/2))
        k4 = func(xi+h,yi+k3*h)
        
        y_new = yi + (h/6)*(k1+2*k2+2*k3+k4)
        x_new = xi + h

        yi = y_new
        xi = x_new

        x_vals.append(xi)
        y_vals.append(yi)

    return np.array(x_vals), np.array(y_vals)
    
```

```python
x_vals, y_vals = RK4(xi=0,yi=0)
plt.plot(x_vals,np.tan(x_vals),label = "tan(x)")
plt.plot(x_vals,y_vals, label="Euler approx")
plt.legend()
```
